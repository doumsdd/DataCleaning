```python
import pandas as pd
import numpy as np

# Load the data with correct delimiter
df = pd.read_csv('/content/openaq.csv', sep=';', error_bad_lines=False)

# Display the first few rows of the dataframe
df.head()
```

on prend uniquement les strings dans la colonne 'last updated' afin de mieux retrouver la date

```python
# Filter the data to include only rows where 'Last Updated' is a string
df = df[df['Last Updated'].apply(lambda x: isinstance(x, str))]

# Display the first few rows of the filtered dataframe
df.head()
```

seulement ceux qui ont Ã©tÃ©s update le '2021-08-06'

```python
# Filter the data to include only rows where 'Last Updated' contains '2021-08-06'
df = df[df['Last Updated'].str.contains('2021-08-06')]

# Display the first few rows of the filtered dataframe
df
```

on liste les polluants

```python
# Check the unique units used for each pollutant
df.groupby('Pollutant')['Unit'].unique()
```

Âµg/mÂ³ est l'unitÃ© la plus rÃ©currente on va donc prendre uniquement les lignes avec cette unitÃ© la et drop les autres .

```python
# Identify the most frequent unit
most_frequent_unit = df['Unit'].value_counts().idxmax()

# Filter rows where 'Unit' is the most frequent unit
df = df[df['Unit'] == most_frequent_unit]

df
```

j'ecourte les valeurs de la colonne "value" a 1 chiffre aprÃ©s la dÃ©cimale .

```python
df['Value'] = df['Value'].astype(float).round(1)
df
```

je sÃ©pare longitude et latitude , j'ecourte les digits a 1 pour chaque . je les rÃ©attribue sur coordonÃ©es pour une meilleure visibilitÃ©

```python
# Split the 'Coordinates' into latitude and longitude
df[['Latitude', 'Longitude']] = df['Coordinates'].str.split(',', expand=True)

# Convert to float and round to 1 decimal place
df['Latitude'] = df['Latitude'].astype(float).round(1)
df['Longitude'] = df['Longitude'].astype(float).round(1)

# Combine back into 'Coordinates'
df['Coordinates'] = df['Latitude'].astype(str) + ',' + df['Longitude'].astype(str)

df
```

je retire le pollutant le moins frÃ©quent

```python
# Identify the least frequent pollutant
least_frequent_pollutant = df['Pollutant'].value_counts().idxmin()

# Filter rows where 'Pollutant' is not the least frequent pollutant
df = df[df['Pollutant'] != least_frequent_pollutant]

df
```

j'enleve latitude et longitude et je drop les NaN

```python
# Drop 'Latitude' and 'Longitude' columns
df = df.drop(['Latitude', 'Longitude'], axis=1)

 # Drop rows with NaN values in any other columns
df.dropna(inplace=True)

# Check for any remaining null values
null_values_after = df.isnull().sum()

null_values_after
```
